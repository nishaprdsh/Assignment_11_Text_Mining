## Extrcating Reviews from Flipkart website-BeutifulSoap

```python
import requests   # Importing requests to extract content from a url
from bs4 import BeautifulSoup  # Beautifulsoup is for web scrapping...used to scrap specific content 
import re 
```

```python
url='https://www.flipkart.com/puma-magda-running-shoes-women/p/itmcf11456b079aa?pid=SHOG5KWEFZFGUXDK&lid=LSTSHOG5KWEFZFGUXDKUHGGDM&marketplace=FLIPKART&store=osp%2Fiko&srno=b_1_3&otracker=clp_omu_Women%2527s%2BFashion_7_3.dealCard.OMU_indias-fashion-capital-store_indias-fashion-capital-store_FDHBS8WZY0XT_2&otracker1=clp_omu_PINNED_neo%2Fmerchandising_Women%2527s%2BFashion_NA_dealCard_cc_7_NA_view-all_2&fm=organic&iid=b44013be-287e-4924-bbc1-161862cbc181.SHOG5KWEFZFGUXDK.SEARCH&ppt=browse&ppn=browse'
```

```python
page=requests.get(url) # making a get request
```

```python
page # check status code for response received
# success code - 200
```

```python
page.text
```

```python
soup=BeautifulSoup(page.text,'html.parser') # Parsing the HTML
print(soup.prettify())
```

```python
reviews=soup.find_all('div',class_='_6K-7Co')
reviews
```

```python
review=[]
for item in reviews:
    item=item.get_text.replace('\n','')
    item=item.strip(' ')
    review.append(item)
```

```python
review
```

```python
ratings=soup.find_all('div',class_='_3LWZlK _1BLPMq _3B8WaH')
ratings
```

```python
rating=[]
for item in ratings:
    item=item.get_text().replace('\n','')
    item=item.strip(' ')
    rating.append(item)
```
